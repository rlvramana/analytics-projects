```python
from datasets import load_from_disk
tokenized_dataset = load_from_disk("tokenized_dataset/")
```

```python
from transformers import BertForSequenceClassification

# Load a pre-trained BERT model with a classification head on top
model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=2,     # Binary classification: 0 or 1
)
```

```python
#  ── Imports ───────────────────────────────────────────────────────
import torch
# tell Torch-Dynamo to fall back to normal eager mode if something
# (e.g. the Trainer) calls torch.compile by accident
import torch._dynamo
torch._dynamo.config.suppress_errors = True

from transformers import TrainingArguments, Trainer, EarlyStoppingCallback
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

#  ── TrainingArguments ─────────────────────────────────────────────
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=10,

    # safe batch that fits 24 GB and still saves time
    per_device_train_batch_size=12,          # 16 could be too tight
    per_device_eval_batch_size=24,
    gradient_accumulation_steps=3,           # 12 × 3 = 36 ≈ 32

    learning_rate=2e-5,
    warmup_ratio=0.06,
    bf16=True,                               # half precision on Apple silicon
    dataloader_pin_memory=False,
    dataloader_num_workers=12,
    group_by_length=True,

    eval_strategy="epoch",                   # correct name for your version
    save_strategy="epoch",
    logging_strategy="steps",
    logging_steps=200,

    torch_compile=False,                     # make absolutely sure compile is off

    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    greater_is_better=True,
    weight_decay=0.01,
    save_total_limit=2,
    seed=42,
)

#  ── Early stopping ───────────────────────────────────────────────
callbacks = [EarlyStoppingCallback(early_stopping_patience=2)]

#  ── Metrics ───────────────────────────────────────────────────────
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average="binary", zero_division=0
    )
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc, "precision": precision, "recall": recall, "f1": f1}

#  ── Trainer ───────────────────────────────────────────────────────
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["val"],
    compute_metrics=compute_metrics,
    callbacks=callbacks,
)
```

```python
trainer.train()
```

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
trainer.save_model('./results/final_model')
tokenizer.save_pretrained('./results/final_model')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the logs from trainer
log_history = trainer.state.log_history
df = pd.DataFrame(log_history)

# Check which columns are available
print("Columns in log history:", df.columns)

# Plot only if 'loss' is present
if 'loss' in df.columns and 'step' in df.columns:
    df_loss = df[df['loss'].notnull()]
    plt.plot(df_loss['step'], df_loss['loss'])
    plt.title("Training Loss over Steps")
    plt.xlabel("Steps")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.show()
else:
    print("No 'loss' or 'step' data found in the log history.")
```

```python
# Plot validation loss
df_val_loss = df[df['eval_loss'].notnull()]
plt.plot(df_val_loss['step'], df_val_loss['eval_loss'], label='Validation Loss', color='orange')
plt.title("Training & Validation Loss over Steps")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()
```

```python
eval_results = trainer.evaluate()
print("Final Evaluation Metrics:")
print(eval_results)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Evaluate on test set (assumes test set is available)
predictions = trainer.predict(tokenized_dataset['test'])
y_true = predictions.label_ids
y_pred = predictions.predictions.argmax(axis=1)

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-Fashion', 'Fashion'])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```
